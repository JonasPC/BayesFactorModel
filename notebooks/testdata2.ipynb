```python
import pandas as pd
import numpy as np
from scipy import linalg
import seaborn as sbn
import matplotlib.pyplot as plt
```

```python
from notebookutils import root_dir, corr_plot; root_dir()
from model.utils import read_clean_kv17
```

```python
np.random.seed(30)
```

### Loading in data

```python
data = read_clean_kv17(drop_party_key = True)
```

```python
corr = np.matrix(data.corr())
corr_plot(data)
```

### Simulating data

```python
L, U = linalg.cholesky(corr, lower=True), linalg.cholesky(corr, lower=False)
```

```python
d, n = 15, 1200
```

```python
X = np.random.normal(size=(d, n))
```

```python
Z = np.dot(L, X)
sim_data = pd.DataFrame(Z.T, columns = data.columns)
```

```python
corr_plot(sim_data)
```
